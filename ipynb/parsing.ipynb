```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
distance = []
time = []
filename = 'PI2'
with open( filename + '.txt', encoding='UTF-8') as file:
    for line in file:
        # time.append(float(line.split()[0][6:8]))
        distance.append(float(line.split()[1].split(',')[0]))

x = [i for i in range(0,len(distance))]
```

```python
count = 0
for i in distance:
    if (i < 10.05) and (i > 9.95):
        print(f" Distance : {i} , when time is : {x[count]} => {float(x[count]) * 0.2}")
    
    count = count + 1
```

```python
y = np.polyfit(x, distance, 50)
print(y)
y = np.poly1d(y)
print(y)
```

```python
plt.plot(x, distance)
plt.plot(x, y(x))
plt.axhline(y=10)
plt.savefig(filename + '_time.png')
plt.show()
```
